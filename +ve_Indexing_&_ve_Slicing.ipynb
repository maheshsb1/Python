```python
var_str_yeh= "Yeh Tera Ghar Yeh Mera Ghar"
print(var_str_yeh)
print(var_str_yeh[26:22:-1])
print(var_str_yeh[21:17:-1])
print(var_str_yeh[16:13:-1])
print(var_str_yeh[12:8:-1])
print(var_str_yeh[7:3:-1])
print(var_str_yeh[2::-1])
print(var_str_yeh[26::-2])
```
