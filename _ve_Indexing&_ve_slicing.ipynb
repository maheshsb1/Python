```python
var_str= "Yeh Tera Ghar Yeh Mera Ghar"
print(var_str)
print(var_str[::-1])
print(var_str[-1::-1])
print(var_str[:-len(var_str):-1])#doubt
print(var_str[-1:-5:-1])
print(var_str[-6:-10:-1])
print(var_str[-11:-14:-1])
print(var_str[-15:-19:-1])
print(var_str[-20:-24:-1])
print(var_str[-25:-28:-1])
print(var_str[-25::-1])
print(var_str[-1::-2])
print(var_str[::-4])
```
